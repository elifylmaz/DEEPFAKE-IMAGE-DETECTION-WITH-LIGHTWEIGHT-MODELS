```python
import os
import cv2
import numpy as np

# Video dosyalarÄ±ndan kare Ã§Ä±karma ve kaydetme
def extract_frames_from_video_and_save(video_path, save_dir, max_frames=30):
    """
    Bu fonksiyon, video dosyasÄ±ndaki kareleri Ã§Ä±kartÄ±r ve belirtilen dizine kaydeder.
    Kareler video adÄ± ve kare numarasÄ± ile adlandÄ±rÄ±lÄ±r.
    """
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)  # EÄer klasÃ¶r yoksa oluÅturulur
    
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frames = []
    
    for i in range(min(max_frames, total_frames)):
        ret, frame = cap.read()
        if ret:
            frame_resized = cv2.resize(frame, (224, 224))  # BoyutlandÄ±rma
            frame_filename = os.path.join(save_dir, f"{os.path.basename(video_path).split('.')[0]}_frame_{i+1}.jpg")
            cv2.imwrite(frame_filename, frame_resized)  # GÃ¶rseli kaydetme
    
    cap.release()

# Ana veri dizini
data_dir = 'Deepfake'  # BurayÄ± doÄru dizinle deÄiÅtirin

# KlasÃ¶rler
real_dir = os.path.join(data_dir, 'Celeb-real')
synthesis_dir = os.path.join(data_dir, 'Celeb-synthesis')
youtube_real_dir = os.path.join(data_dir, 'YouTube-real')

# Kareleri Ã§Ä±karma ve kaydetme
def process_videos_in_directory(input_dir, output_dir):
    """
    Verilen klasÃ¶rdeki videolarÄ± iÅler ve karelerini belirtilen klasÃ¶re kaydeder.
    """
    for file in os.listdir(input_dir):
        if file.endswith(".mp4"):
            video_path = os.path.join(input_dir, file)
            save_dir = os.path.join(output_dir, file.split('.')[0])  # Video ismiyle alt klasÃ¶r oluÅturulacak
            extract_frames_from_video_and_save(video_path, save_dir)

# GerÃ§ek yÃ¼zler (Celeb-real) iÃ§in kareleri Ã§Ä±karma
process_videos_in_directory(real_dir, os.path.join(data_dir, 'Frames', 'Celeb-real'))

# SentezlenmiÅ yÃ¼zler (Celeb-synthesis) iÃ§in kareleri Ã§Ä±karma
process_videos_in_directory(synthesis_dir, os.path.join(data_dir, 'Frames', 'Celeb-synthesis'))

# YouTube gerÃ§ek yÃ¼zler (YouTube-real) iÃ§in kareleri Ã§Ä±karma
process_videos_in_directory(youtube_real_dir, os.path.join(data_dir, 'Frames', 'YouTube-real'))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
import pandas as pd

# Etiketleri oluÅturma
def create_dataframe(frame_dir, label):
    filepaths = []
    labels = []
    for root, _, files in os.walk(frame_dir):
        for file in files:
            if file.endswith(".jpg"):
                filepaths.append(os.path.join(root, file))
                labels.append(label)
    return pd.DataFrame({'filepath': filepaths, 'label': labels})

# DataFrame'leri birleÅtirme
real_df = create_dataframe(os.path.join(data_dir, 'Frames', 'Celeb-real'), 'real')
synth_df = create_dataframe(os.path.join(data_dir, 'Frames', 'Celeb-synthesis'), 'fake')
youtube_df = create_dataframe(os.path.join(data_dir, 'Frames', 'YouTube-real'), 'real')

full_df = pd.concat([real_df, synth_df, youtube_df], axis=0)

# Veriyi bÃ¶lme
train_df, temp_df = train_test_split(full_df, test_size=0.3, stratify=full_df['label'])
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['label'])

# Data Augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

BATCH_SIZE = 32
IMG_SIZE = (224, 224)

train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col='filepath',
    y_col='label',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

val_generator = val_test_datagen.flow_from_dataframe(
    dataframe=val_df,
    x_col='filepath',
    y_col='label',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

test_generator = val_test_datagen.flow_from_dataframe(
    dataframe=test_df,
    x_col='filepath',
    y_col='label',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)
```

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
import tensorflow as tf  

def create_model():
    base_model = EfficientNetB0(
        weights='imagenet',
        include_top=False,
        input_shape=(224, 224, 3)
    )
    
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(128, activation='relu')(x)
    predictions = Dense(1, activation='sigmoid')(x)
    #bu tarz eklemensi lazÄ±m
    model = Model(inputs=base_model.input, outputs=predictions,device = "cuda)
    
    # Ä°lk katmanlarÄ± dondurma
    for layer in base_model.layers[:-4]:
        layer.trainable = False
        
    return model

model = create_model()
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

checkpoint = ModelCheckpoint(
    'best_model.h5',
    monitor='val_loss',
    save_best_only=True,
    verbose=1
)

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    verbose=1
)

history = model.fit(
    train_generator,
    epochs=20,
    validation_data=val_generator,
    callbacks=[checkpoint, early_stop]
)
```

```python
import matplotlib.pyplot as plt

# Accuracy ve Loss Grafikleri
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Progress')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Progress')
plt.legend()
plt.show()

# Test DeÄerlendirmesi
from sklearn.metrics import classification_report, confusion_matrix

model.load_weights('best_model.h5')
test_preds = model.predict(test_generator)
test_preds_binary = (test_preds > 0.5).astype(int)

print("Test SonuÃ§larÄ±:")
print(classification_report(test_generator.labels, test_preds_binary))
print("Confusion Matrix:")
print(confusion_matrix(test_generator.labels, test_preds_binary))
```

```python
# Modeli kaydetme
model.save('deepfake_detection_model.h5')

# Yeni veriyle tahmin Ã¶rneÄi
def predict_deepfake(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))
    img = img / 255.0
    img = np.expand_dims(img, axis=0)
    
    prediction = model.predict(img)
    return "Sahte" if prediction[0][0] > 0.5 else "GerÃ§ek"

# Ãrnek kullanÄ±m
print(predict_deepfake('test_image.jpg'))
```
